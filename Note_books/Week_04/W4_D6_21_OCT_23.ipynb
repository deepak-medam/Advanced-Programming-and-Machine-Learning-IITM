```python
#Find factorial of the number 6

## 6! FACTORIAL is denoted by 1*2*3*4*5*6 = 720

#num = 6

num = int(input("Enter a number: "))

factorial = 1

# check if the number is negative, positive or zero
if num < 0:
   print("Sorry, factorial does not exist for negative numbers")
elif num == 0:
   print("The factorial of 0 is 1")
else:
   for i in range(1,num + 1):
       factorial = factorial*i
print("The factorial of",num,"is",factorial)
```

#### 6. Python code that copies an array and turns each letter in that array into an uppercase. It does so by looping through each item in the array?

```python
letters = ['a', 'b', 'c', 'd']
print(letters)

upper_letters = []
for letter in letters:
   result = letter.upper()
   upper_letters.append(result)

print(upper_letters)
```

#### 4.Variable-length arguments

## *args in function definitions in python is used to pass a variable number of arguments to a function

It takes at least one argument (arg1) and can take any number of additional arguments (*arg), which are gathered into a tuple.

using * before a parameter in a function allows that parameter to accept multiple values (as a tuple) when the function is called.


```python
# read the value one by one and prints the value
# *args in function definitions in python is used to pass a variable number of arguments to a function
# symbol * to take in a variable number of arguments

def printinfo( arg1, *arg ):
   print(arg1)

   for var in arg:
       print(var)

#printinfo( 10 )
printinfo( 10,20,30 )
```

```python
# * denots number of arguments
def myFun(*argv):
   for arg in argv:
       print (arg)

myFun("hello","guvi","chennai")
```

## **kwargs in function definitions in python is used to pass a keyworded, variable-length argument list

      # keyword - provide name to the arguments
      # ** pass through keyword argument
      # this helps to unpack dictonary

The ** before kwargs in the function definition allows the function to accept a variable number of keyword arguments.

It gathers these keyword arguments into a dictionary named kwargs, where the keys are the argument names and the values are the argument values.


```python
def fun1(**kwargs):
   for k,v in kwargs.items():
       print("%s == %s"%(k,v))

fun1(first = "Hello",mid = "guvi", end = "chennai")
```

## *before a parameter gathers additional positional arguments into a tuple.

## ** before a parameter gathers additional keyword arguments into a dictionary.


```python
def print_keyword_args(**kwargs):
    # kwargs is a dict of the keyword args passed to the function
       for key, value in kwargs.items():
           print( "%s = %s" % (key, value))

#call the function
print_keyword_args(first_name="John", last_name="Doe")
```
