```python
# define a function palindrome
# get the list of all the inputs values, find if each word is a palindrome or not
# display the result in 2 different list
def pal(a):
   pal = []
   nonpal = []

   for i in a:
       i1 = i[::-1]
       if(i == i1):
           pal.append(i)
       else:
           nonpal.append(i)

   print("pal : ", pal)
   print("nonpal :",nonpal)


# call the function
#a = ["greet","wow","learning","madam","awesome"]

# pass the value with space "greet" "wow" "learning" "madam" "awesome"
lst  = input("enter the list of words: ")
a  = lst.split()
pal(a)
```

```python
def prime(num):
   # prime number is always greater than 1
   if(num>1):

       for i in range(2, num):
           if(num%i == 0):
               print('It is NOT a prime number')
               break
       else:
           print('It is a prime number')
   else:
       print('It is NOT a prime number')

#call thefunction
# Enter a positive integer

num = int(input('Enter a positive integer :  '))

prime(num)
```

#### Write a function to find the factorial, allow only positive integer alone

```python
# Write a function to find the factorial, allow only positive integer alone
## 6! FACTORIAL is denoted by 1*2*3*4*5*6 = 720
#num = 6

num = int(input("Enter a number: "))

def fact(num):
   factorial = 1

   # check if the number is negative, positive or zero
   if num < 0:
       print("Sorry, factorial does not exist for negative numbers")
   elif num == 0:
       print("The factorial of 0 is 1")
   else:
       for i in range(1,num + 1):
           factorial = factorial*i
   print("The factorial of",num,"is",factorial)


#call the function fact

fact(num)
```

#### factorial using while loop

```python
# factorial using while loop

#### 6! FACTORIAL is denoted by 1*2*3*4*5*6 = 720
num = int(input("enter a number: "))
def fact_while(num):
   fac = 1
   i = 1

   while i <= num:
       fac = fac * i
       i = i + 1

   print("factorial of ", num, " is ", fac)

fact_while(num)
```

##### global keyword we are telling that a is globally defined, so when we are changing as value inside the function it is actually changing for the a outside of the function also.


```python
# end to end example for local and gloabl variable inside function
a = 1
 # Uses global because there is no local 'a'

def f():
   print( 'Inside f() : ', a )
 # Variable 'a' is redefined as a local
def g():
   a = 2 # modified the value of a
   print('Inside g() : ',a )
 # Uses global keyword to modify global 'a'
def h():
   global a
   a = 3
   print('Inside h() : ',a )
 # Global scope
print('global : ',a )
f()
print('global : ',a )
g()
print('global : ',a )
h()
print('global : ',a )

print("value:",a)
```
