### Ex:1 Create a rank column and pass the returned rank series

**The rank() function is used to compute the rank of each value in the "Maths" column. It assigns a unique rank to each value based on their relative positions in the column, with ties being handled based on the method used (default is "average").**

```python
scores_df["Maths_Rank"] = scores_df["Maths"].rank()
scores_df
```

```python
scores_df.sort_values(by='Maths',ascending=0,inplace = True)
scores_df          # descending order
```

```python
scores_df["Maths_Rank1"] = scores_df["Maths"].rank()
scores_df
```

### Ex:2 Create a rank column and pass the returned rank series (rank in descending order)

```python
scores_df["Maths_Rank2"] = scores_df["Maths"].rank(ascending=0)
scores_df
```

### Ex:3 Create a rank column using 'min' method

```python
scores_df["Maths_Rank3"] = scores_df["Maths"].rank(ascending=0,method='min')
scores_df
```

### Ex:4 Create a rank column using 'dense' method

**The rank() function is used to compute the ranks. The additional parameters are:**

ascending=0: This parameter specifies that the ranking should be in descending order. In other words, the highest values will have the lowest rank (1 for the highest value). method='dense': This parameter specifies the method for handling ties (when multiple values are the same). The 'dense' method assigns the same rank to tied values and then increments the rank by 1 for the next distinct value.


```python
scores_df["Maths_Rank4"] = scores_df["Maths"].rank(ascending=0,method='dense')
scores_df
```

```python
image_12 = Image.open("/content/drive/MyDrive/2023-24/courses/Advanced-Programming-and-Machine-Learning-IITM/Images/Week_06/10.12.png")
display.display(image_12)
```

**The pd.melt() function in the Pandas library is used to transform or reshape a DataFrame from a wide format to a long format. It's a useful tool when you want to restructure your data to make it more suitable for certain types of analysis, visualization, or modeling. Here are some common use cases for using pd.melt().**

```python
data = pd.DataFrame({
   'Name': ['Alice', 'Bob', 'Charlie'],
   'Math_Score': [90, 85, 78],
   'English_Score': [88, 76, 92],
   'History_Score': [75, 82, 89]
})
data
```

**we've melted the original DataFrame, converting it from a wide format with subject-specific columns to a long format where each row represents a student, their subject, and their score. This long format can be more suitable for various types of analysis and visualization.**

```python
melted_data = pd.melt(data, id_vars=['Name'], var_name='Subject', value_name='Score')
melted_data
```

## VII. Pivot Table

```python
#Create a DataFrame

d = {
   'Name':['Jony','Anil','Sunil','Jony','Anil','Rocky',
           'Jony','Anil','Sunil','Jony','Anil','Rocky'],
   'Exam_Months':['3_Months','3_Months','3_Months','3_Months','3_Months','3_Months',
           '6_Months','6_Months','6_Months','6_Months','6_Months','6_Months'],

   'Subject':['Python','Python','SQL','SQL','ML','ML',
              'Python','Python','SQL','SQL','ML','ML'],
  'Score':[62,47,55,74,31,77,85,63,42,67,89,81]}
df = pd.DataFrame(d,columns=['Name','Exam_Months','Subject','Score'])
df
```

**The pivot() function in Pandas is used to reshape a DataFrame from a long or tidy format into a wide format, where the rows and columns are reorganized to create a new structure of the data. It is essentially the opposite of the melt() function. The primary purpose of pivot() is to pivot data for better analysis, reporting, or presentation.**

```python
image_13 = Image.open("/content/drive/MyDrive/2023-24/courses/Advanced-Programming-and-Machine-Learning-IITM/Images/Week_06/10.13.png")
display.display(image_13)
```
