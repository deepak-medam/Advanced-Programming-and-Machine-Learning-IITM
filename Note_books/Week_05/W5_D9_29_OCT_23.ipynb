## VI. Creating dataframe using Dictionary (.from_dict)

The default manner to create a DataFrame from python is to use a list of dictionaries. In this case each dictionary key is used for the column headings. A default index will be created automatically:

```python
gg= ({"Location":"Bangalore","Jan":150,"Feb":80},
    {"Location":"Chennai","Jan":200,"Feb":70})
Df = pd.DataFrame(gg)
Df
```

```python
Guvi = ({"Location":["Bangalore","Chennai"],
      "Jan":[150,80],
      "Feb":[200,70]})

df = pd.DataFrame.from_dict(Guvi)
df
```

## VII. Creating dataframe using List

The other option for creating your DataFrames from python is to include the data in a list structure.

The first approach is to use a row oriented approach using pandas from_records . This approach is similar to the dictionary approach but you need to explicitly call out the column labels.

```python
Guvi = [("Bangalore",150,80),
     ("Chennai",200,70)]
Labels = ["Location","Jan","Feb"]
df = pd.DataFrame.from_records(Guvi,columns=Labels)
df
```

```python
Guvi = [("Location", ["Bangalore","Chennai"]),("Jan",[150,200]),("Feb",[80,70])]

# Extract the column names and data from the list of tuples
columns = [item[0] for item in Guvi]
data = [item[1] for item in Guvi]

# Create the DataFrame
df1 = pd.DataFrame(dict(zip(columns, data)))
print(df1)
```
