## 2.Built-in Dictionary methods

A.    dic.clear()      =It is used to delete all the items of the dictionary.

B.    dict.copy()        =It returns a shallow copy of the dictionary.

C.    dict.fromkeys()    =Create a new dictionary from the iterable with the values equal to value.

D.  dict.get(key[, value])        =It is used to get the value specified for the passed key.

F.    dict.items()    =It returns all the key-value pairs as a tuple.

I.    dict.keys()     =It returns all the keys of the dictionary.

J.    dict.setdefault()    =It is used to set the key to the default value if the key is not specified in the dictionary

K.    dict.update()     =It updates the dictionary by adding the key-value pair of dict2 to this dictionary.

L.    dict.values()    =It returns all the values of the dictionary.

### 2.B) dict.copy():

```python
Dictionaries = {1: "Hi", 2: "Guvi"}
new = Dictionaries.copy()

print('Orignal: ', Dictionaries)
print('New: ', new)
```

### 2.C) dict.fromkeys():

```python
# seq is a tuple
seq = ('name', 'age', 'gender')
# tuple is type casted into dict and all the elements are converted into key's. So they key will not have any values.
dict = dict.fromkeys(seq)

print ("New Dictionary : %s" %  str(dict))
# assign a default value as 10 for all the keys
dict = dict.fromkeys(seq, 10)
print ("New Dictionary with value: %s" %  str(dict))
```

### 2.G) dict.setdefault():


```python
person = {'name': 'Phill'}
print(person.keys())
print(person.values())
print(person.items())
```

```python
# key is not in the dictionary
salary = person.setdefault('salary' )
print('person = ',person)
print('salary = ',salary)
```

```python
person = {'name': 'Phill'}
# key is not in the dictionary, add salarya nd give default value as 5000
salary = person.setdefault('salary' , 5000)
print('person = ',person)
print('salary = ',salary)
```

### 2.H) dict.update():

```python
# create a dict key - fruit name, value is price
prices = {'apple': 0.40, 'orange': 0.35, 'banana': 0.25}

# Get the key and values for each items
for k, v in prices.items():
   #for each key value, get its respective value and
   prices[k] = round(v * 0.9, 2)  # Apply a 10% discount
prices
```

## Nested For
    for [first iterating variable] in [outer loop]: # Outer loop
      [do something]  # Optional
      for [second iterating variable] in [nested loop]:   # Nested loop
          [do something]  


```python
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
   for y in fruits:
       print(x, y)
```

```python
num_list = [1, 2, 3]
alpha_list = ['a', 'b', 'c']

for number in num_list:
   print(number)
   for letter in alpha_list:
       print(letter)
```

```python
list_of_lists = [['hello', 'guvi','datascience'],[0, 1, 2],[9.9, 8.8, 7.7]]

for list in list_of_lists:
   for item in list:
       print(item)
```
