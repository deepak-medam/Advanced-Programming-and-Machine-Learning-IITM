10.Write a python program that takes a list ['red','black','blue'] and prints each element in the list using whle loop

```python
color_lis = ['red', 'black', 'blue']
curr_idx = 0
while curr_idx < len(color_lis):
  print(color_lis[curr_idx])
  curr_idx += 1
```

12.Write a Python Program to find the LCM of two numbers?

```python
def lcm(x, y):
  if x < y:
    x, y = y, x

  while y % x != 0:
    x, y = y, x % y

  return x

x = int(input("Enter your first integer: "))
y = int(input("Enter your second integer: "))
lcm(x,y)
```

13.Python Program to find the sum of series: 1 + 1/2 + 1/3 + â€¦.. + 1/N.

```python
N = int(input("Enter the value of N: "))
sum = 1

for i in range(1, N + 1):
    sum += 1 / i

# Print the result
print("Sum of the series is:", sum)
```

## List Comprehensions

1.
  
  i. Write a Python program to get the squares of numbers present in the range of (1-10)
* output squares = [0, 1, 4, 9, 16, 25, 36, 49, 64, 81, 100]


```python
output_square = []
for num in range(1, 11):
  output_square.append(num**2)

print(output_square)
```


ii. For the same question get the desired output using List Comprehension


```python
square = [x**2 for x in range(1,11)]
print(square)
```

2.Convert a list of integers to a list of strings, using List comprehension

```python
num_list = [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]
num_list = [str(num) for num in num_list]
print(num_list)
```

3.Create tuples from two lists using List Comprehensions

```python
list1 = [1, 2, 3, 4, 5]
list2 = ['a', 'b', 'c', 'd', 'e']

result = [(x, y) for x, y in zip(list1, list2)]
print(result)
```
